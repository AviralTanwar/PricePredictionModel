```python
# Filter the DataFrame based on the Price condition (<10K and >10K)
df_lt_10k = for_graphs[for_graphs['Price'] < 10000]
df_gt_10k = for_graphs[for_graphs['Price'] > 10000]

# Calculate the counts for each category within the filtered DataFrames
counts_lt_10k = df_lt_10k['type'].value_counts()
counts_gt_10k = df_gt_10k['type'].value_counts()

# Prepare the data and ingredients for pie charts
data_lt_10k = [counts_lt_10k['Smartwatches'], counts_lt_10k['Smartphones'], counts_lt_10k['Smarttablets']]
ingredients = ['Smartwatches', 'Smartphones', 'Smarttablets']

data_gt_10k = [counts_gt_10k['Smartwatches'], counts_gt_10k['Smartphones'], counts_gt_10k['Smarttablets']]

def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return f"{pct:.1f}%\n({absolute:d} devices)"


# Create the first pie chart for Price < 10K
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

wedges, texts, autotexts = ax.pie(data_lt_10k, autopct=lambda pct: func(pct, data_lt_10k),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Devices",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Smart Devices Prices (< 10K)")

plt.show()

# Create the second pie chart for Price > 10K
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

wedges, texts, autotexts = ax.pie(data_gt_10k, autopct=lambda pct: func(pct, data_gt_10k),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Devices",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")

ax.set_title("Smart Devices Prices (> 10K)")

plt.show()
```
